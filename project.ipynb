# Formatting Data

```python
house_Vancouver_price = houseVancouver[['Date',
                                    'Composite_Benchmark_SA', 
                                    'Single_Family_Benchmark_SA',
                                    'One_Storey_Benchmark_SA',
                                    'Two_Storey_Benchmark_SA',
                                    'Townhouse_Benchmark_SA',
                                    'Apartment_Benchmark_SA']]

house_Vancouver_price.set_index('Date',inplace = True)
house_Vancouver_perc = houseVancouver[['Date',
                                  'Composite_HPI_SA',
                                  'Single_Family_HPI_SA',
                                  'One_Storey_HPI_SA',
                                  'Two_Storey_HPI_SA',
                                  'Townhouse_HPI_SA',
                                  'Apartment_HPI_SA']]
house_Vancouver_perc.set_index('Date', inplace = True)
```

## Bar plot for the housing price and percentage 

```python
house_Vancouver_price.plot.bar(figsize = (50,30))
house_Vancouver_perc.plot.bar(figsize = (50,30))

plt.show()
```

## Bar plot for the number of immigrants

```python
numberOfImmigrants = censusImm[['Census years', 'Number']]
numberOfImmigrants.set_index('Census years', inplace = True)
numberOfImmigrants.plot.bar(figsize = (50,30))
plt.show()
```

# Analysis 

### We can see from the plot that, the housing prices are increasing relative to the growing number of immigrants to Canada.
