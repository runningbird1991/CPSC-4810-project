# Formatting Data

```python
house_Vancouver_price = houseVancouver[['Date',
                                    'Composite_Benchmark_SA', 
                                    'Single_Family_Benchmark_SA',
                                    'One_Storey_Benchmark_SA',
                                    'Two_Storey_Benchmark_SA',
                                    'Townhouse_Benchmark_SA',
                                    'Apartment_Benchmark_SA']]

house_Vancouver_price.set_index('Date',inplace = True)
house_Vancouver_perc = houseVancouver[['Date',
                                  'Composite_HPI_SA',
                                  'Single_Family_HPI_SA',
                                  'One_Storey_HPI_SA',
                                  'Two_Storey_HPI_SA',
                                  'Townhouse_HPI_SA',
                                  'Apartment_HPI_SA']]
house_Vancouver_perc.set_index('Date', inplace = True)
```

## Formating the Vancouver housing price Dataset

```python
house_Vancouver_price_DateColumn = house_Vancouver_price.reset_index()
new_composite = house_Vancouver_price_DateColumn[(house_Vancouver_price_DateColumn['Date'] == pd.Timestamp(2006, 1, 1, 0)) | (house_Vancouver_price_DateColumn['Date'] == pd.Timestamp(2011, 1, 1, 0)) | (house_Vancouver_price_DateColumn['Date'] == pd.Timestamp(2016, 1, 1, 0)) | (house_Vancouver_price_DateColumn['Date'] == pd.Timestamp(2021, 1, 1, 0))]
new_composite.reset_index(inplace = True, drop = True)
print()
new_composite.iloc[0,0] = 2006
new_composite.iloc[1,0] = 2011
new_composite.iloc[2,0] = 2016
new_composite.iloc[3,0] = 2021
display(new_composite)
```

## Correlatoin tables

```python
corr = join_data_set.corr()
correlation_data = corr['Number']
```

## Correlation Plots

```python
positive_correlation = correlation_data[correlation_data >= 0]
negative_correlation = correlation_data[correlation_data < 0]
positive_correlation.index
```

## Positve Correlation

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = positive_correlation.index
y = positive_correlation.values
ax.barh(x,y)
plt.xticks([0,0.3,0.7,1],rotation=45)
plt.grid(axis='x')
sns.despine(left=True, bottom=False)
plt.show()
```

## Negative Correlation

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = negative_correlation.index
y = negative_correlation.values
ax.barh(x,y)
plt.xticks([0,-0.3,-0.7,-1],rotation=45)
plt.grid(axis='x')
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
sns.despine(left=True, bottom=False)
plt.show()
```

# Correlation Analysis     

### The following table shows that a correlation can be weak, moderate, or strong based on the value of r. 

     weak correlation              moderate correlation                 strong correlation   


       0 < |r| < 0.3                 0.3 < |r| < 0.7                       |r| > 0.7

From the above correlation plots, we can conclude the followings,

1: The number of the immigration has strong correlation with the increase in housing price in Vancouver

2: The number of the immigration does not have a weak or negative correlation with the increase in housing price in Vancouver
